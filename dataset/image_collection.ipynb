```python
import numpy as np
import pandas as pd

import requests
```

```python
# read data and save as df
df = pd.read_csv('memes_data.tsv', sep='\t')
```

```python
# drop duplicates from dataset
df1 = df.drop_duplicates(subset=['HashId'])
```

```python
df1
```

```python
# collecting and saving images from web
for i, row in df1.iterrows():
    url = 'https:' + row.ImageURL
    r = requests.get(url)

    with open('D:/lh_final_data_images/img' + str(i) +'.jpg', 'wb') as outfile:
        outfile.write(r.content)
```
